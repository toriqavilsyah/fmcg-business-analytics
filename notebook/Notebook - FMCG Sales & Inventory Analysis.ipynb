# Distribution Audit

```python
# Uji Distribusi

from scipy.stats import kstest
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:

    # Standarisasi (Z-score)
    z_score = (df[col] - df[col].mean()) / df[col].std()

    # Drop NA jika ada
    z_score = z_score.dropna()

    # Uji KS
    stat, p = kstest(z_score, 'norm')

    print(f"Kolom: {col}")
    print(f"  Statistik uji : {stat:.4f}")
    print(f"  p-value       : {p:.6f}")

    if p > 0.05:
        print("Data berdistribusi normal\n")
    else:
        print("Data tidak berdistribusi normal\n")
```

- Hasil uji Kolmogorov–Smirnov menunjukkan bahwa seluruh variabel numerik memiliki p-value lebih kecil dari 0.05, sehingga data tidak berdistribusi normal.

- Hal ini mengindikasikan bahwa distribusi data cenderung skewed atau memiliki pola distribusi non-normal, sehingga pada analisis selanjutnya metode non-parametrik atau transformasi data dapat dipertimbangkan untuk memperoleh hasil analisis yang lebih robust.

## Handling Outlier

Penanganan outlier dilakukan menggunakan metode **capping (winsorization)** dengan pendekatan **Interquartile Range (IQR)**. Metode ini tidak menghapus data yang dianggap outlier, tetapi membatasi nilai ekstrem agar berada dalam batas distribusi yang lebih wajar.

Pendekatan ini dipilih untuk menjaga jumlah observasi tetap utuh sekaligus mengurangi pengaruh nilai ekstrem yang dapat mendistorsi analisis statistik.

```python
# Handling Outlier dengan metode Capping (Winsorization) menggunakan IQR Method
num_cols = df_transform.select_dtypes(include=['int64','float64']).columns
num_cols = [col for col in num_cols if df_transform[col].nunique() > 2]

for col in num_cols:

    Q1 = df_transform[col].quantile(0.25)
    Q3 = df_transform[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_transform[col] = df_transform[col].clip(lower, upper)
```

## Time Series Analysis

Metode *Time Series Analysis* digunakan untuk menganalisis perubahan penjualan dari waktu ke waktu berdasarkan variabel waktu (`date`). Analisis ini bertujuan untuk mengidentifikasi pola penjualan yang terjadi sepanjang periode pengamatan.

Melalui pendekatan ini, beberapa aspek utama yang dapat dianalisis meliputi:

- **Tren penjualan**, yaitu kecenderungan peningkatan, penurunan, atau kestabilan penjualan dari waktu ke waktu.
- **Fluktuasi permintaan**, yaitu variasi atau perubahan jumlah penjualan yang terjadi dalam periode tertentu.
- **Pola musiman (seasonality)**, yaitu pola penjualan yang berulang pada periode waktu tertentu, seperti bulanan atau tahunan.

Analisis ini membantu memahami dinamika permintaan produk serta menjadi dasar dalam perencanaan stok dan distribusi yang lebih efektif.

```python
# membuat dataset time series (sales per hari)
sales_time = df_final.groupby('date')['units_sold'].sum().reset_index()

# urutkan berdasarkan tanggal
sales_time = sales_time.sort_values('date')

# membuat moving average
sales_time['MA_7'] = sales_time['units_sold'].rolling(window=7).mean()
sales_time['MA_30'] = sales_time['units_sold'].rolling(window=30).mean()

# visualisasi trend penjualan
plt.figure(figsize=(12,6))

plt.plot(sales_time['date'], sales_time['units_sold'], label='Daily Sales', alpha=0.5)

plt.plot(sales_time['date'], sales_time['MA_7'], label='7-Day Moving Average', linewidth=2)

plt.plot(sales_time['date'], sales_time['MA_30'], label='30-Day Moving Average', linewidth=2)

plt.title('Trend Penjualan Produk dari Waktu ke Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Units Sold')
plt.legend()

plt.show()
```

```python
# Analisis Seasonality (Pola Musiman)
df_final.groupby('month')['units_sold'].sum().plot(
    kind='line',
    figsize=(10,5),
    color='skyblue',
    title='Pola Musiman Penjualan Berdasarkan Bulan'
)

plt.xlabel('Bulan')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=0)

plt.show()
```

### Interpretasi Time Series dan Pola Musiman

Berdasarkan visualisasi *time series* menggunakan **moving average (MA-7 dan MA-30)**, terlihat bahwa penjualan mengalami tren peningkatan dari awal periode hingga sekitar pertengahan 2023 sebelum kemudian mengalami penurunan dan fluktuasi pada periode berikutnya. Moving average membantu memperjelas tren utama dengan mengurangi fluktuasi harian pada data penjualan.

Analisis **pola musiman bulanan** menunjukkan bahwa penjualan cenderung meningkat dari awal tahun hingga mencapai puncaknya sekitar **bulan Juli**, kemudian mengalami penurunan secara bertahap hingga akhir tahun. Pola ini mengindikasikan adanya permintaan yang lebih tinggi pada pertengahan tahun dibandingkan periode lainnya.

Metode **moving average dan analisis seasonality tidak memerlukan asumsi distribusi normal**, karena teknik ini hanya digunakan untuk menghaluskan fluktuasi data serta mengidentifikasi pola tren dan musiman dalam data deret waktu. Oleh karena itu, analisis lebih berfokus pada **pola temporal (trend dan seasonality)** daripada bentuk distribusi statistik dari data penjualan.

```python
# Persiapan Data untuk Forecasting
# membutuhkan dua kolom dengan nama khusus yaitu : ds (date) dan y (nilai yang diprediksi (units_sold)

forecast_data = sales_time[['date','units_sold']].rename(
    columns={
        'date':'ds',
        'units_sold':'y'
    }
)
```

```python
# import library prophet
from prophet import Prophet

# Membuat dan melatih model
model = Prophet()
model.fit(forecast_data)

# Membuat prediksi 90 hari ke depan
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)

# Visualisasi hasil forecast
model.plot(forecast)

plt.title("Forecast Penjualan Produk")
plt.xlabel("Tanggal")
plt.ylabel("Units Sold")

plt.show()
```

### Interpretasi Hasil Forecasting

1. Berdasarkan hasil forecasting menggunakan model Prophet, prediksi penjualan dipengaruhi oleh kombinasi tren jangka panjang serta pola musiman yang terdapat dalam data historis.
2. Komponen **trend** menunjukkan bahwa penjualan mengalami peningkatan yang signifikan dari awal periode hingga sekitar pertengahan 2023, kemudian mengalami penurunan bertahap dan cenderung stabil pada periode berikutnya.
3. Komponen **weekly seasonality** menunjukkan adanya variasi penjualan berdasarkan hari dalam satu minggu, di mana penjualan cenderung lebih tinggi pada awal minggu dan relatif lebih rendah menjelang akhir minggu.
4. Sementara itu, **yearly seasonality** menunjukkan adanya pola musiman tahunan, di mana penjualan cenderung meningkat menuju pertengahan tahun dan menurun kembali pada akhir tahun.
5. Dengan mempertimbangkan tren dan pola musiman tersebut, model memprediksi bahwa **penjualan pada periode mendatang akan mengikuti pola historis yang serupa dengan fluktuasi yang relatif stabil**.

```python
# Evaluasi Akurasi Model Prophet

from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Split data (90 hari terakhir sebagai test)
train = forecast_data.iloc[:-90]
test = forecast_data.iloc[-90:]

# Train model
model = Prophet()
model.fit(train)

# Buat prediksi
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)

# Ambil prediksi untuk periode test
forecast_test = forecast[['ds','yhat']].tail(90)

# Gabungkan dengan data aktual
evaluation = test.merge(forecast_test, on='ds')

# Hitung metrik akurasi
mae = mean_absolute_error(evaluation['y'], evaluation['yhat'])
rmse = np.sqrt(mean_squared_error(evaluation['y'], evaluation['yhat']))
mape = np.mean(np.abs((evaluation['y'] - evaluation['yhat']) / evaluation['y'])) * 100

print("MAE  :", mae)
print("RMSE :", rmse)
print("MAPE :", mape)
```

### Evaluasi Akurasi Model Forecasting

Untuk mengevaluasi performa model forecasting, dilakukan pembagian data menjadi data pelatihan (*training set*) dan data pengujian (*test set*) dengan menggunakan 90 hari terakhir sebagai periode pengujian.

Berdasarkan hasil evaluasi, diperoleh metrik sebagai berikut:

- **MAE (Mean Absolute Error)** : 155.14  
- **RMSE (Root Mean Squared Error)** : 194.93  
- **MAPE (Mean Absolute Percentage Error)** : 4.12%

Nilai **MAPE sebesar 4.12%** menunjukkan bahwa rata-rata kesalahan prediksi model hanya sekitar 4% dari nilai aktual. Berdasarkan standar evaluasi model forecasting, nilai MAPE di bawah 10% dapat dikategorikan sebagai **sangat akurat**.

Dengan demikian, model Prophet mampu menangkap pola historis penjualan dengan baik dan menghasilkan prediksi yang cukup andal untuk digunakan dalam perencanaan stok dan distribusi produk di masa mendatang.

```python
df_final['stock_level'] = pd.qcut(
    df_final['stock_available'],
    q=3,
    labels=['Low Stock','Medium Stock','High Stock']
)
```

## Time Features

- Feature Time Features dibuat dengan **mengekstrak informasi waktu dari variabel date untuk mendukung analisis pola penjualan berbasis waktu**.
- Feature ini digunakan untuk **mengidentifikasi tren penjualan, pola musiman (seasonality), serta fluktuasi permintaan, dan juga digunakan sebagai dasar dalam proses sales forecasting**.

```python
df_final['year'] = df_final['date'].dt.year
df_final['month'] = df_final['date'].dt.month
df_final['week'] = df_final['date'].dt.isocalendar().week
```

- Metrik ini digunakan sebagai indikator utama performa penjualan karena menunjukkan tingkat permintaan pasar terhadap produk.

```python
total_unit_sold = df_final['units_sold'].sum()
```

## Stock Utilization Rate

- **Stock Utilization** mengukur tingkat pemanfaatan stok yang tersedia.  
Metrik ini dihitung sebagai rasio antara jumlah unit yang terjual dengan jumlah stok yang tersedia.
- Nilai ini menunjukkan seberapa besar stok yang berhasil dikonversi menjadi penjualan, sehingga dapat digunakan untuk mengevaluasi efisiensi pengelolaan persediaan produk.

```python
stock_utilization_rate = df_final["units_sold"].sum() / df_final["stock_available"].sum()
```
